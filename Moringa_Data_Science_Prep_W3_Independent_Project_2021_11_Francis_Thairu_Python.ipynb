#Dataset 1 

```python
#we read the Telkom_dataset1 file we uploaded as shown below.


telc_df = pd.read_csv("Telcom_dataset.csv")
telc_df

#we the change the column names to the appropriate names 

telc_df.columns = ['PRODUCT',	'VALUE',	'DATE_TIME',	'CELL_ON_SITE',	'DW_A_NUMBER',	'DW_B_NUMBER',	'COUNTRY_A',	'COUNTRY_B',	'CELL_ID',	'SITE_ID']
telc_df


```

```python
#we find if there are null values and we drop them
telc_df.isnull()
telcdf = telc_df.dropna()
telcdf
```

```python
#we check for duplicates and drop them

telc_df.duplicated()
telcdf.drop_duplicates()
```

```python
#we merge the fist dataset table with the cells geo table
new_df=pd.concat([telcdf, cell_geo_data], axis=1)
new_df
```

```python
new_df.groupby('Cell_ID').count().nlargest(5,'Product')
```

```python
#We find the maximum values with respect to cities

new_df1 = new_df.groupby('City').max('VALUE')
new_df1
```

```python
#we find th maximum value
f1=new_df1['VALUE'].max()
f1
```

```python
#We find the city with the largest value
new_df1.groupby('City').count().nlargest(1,'VALUE')
```
